## Reduce size of an image

```python
from PIL import Image

# Open the original image
original_image = Image.open('lena.png')  # Replace 'original_image.jpg' with your image file

# Resize the image to 50x50 pixels
resized_image = original_image.resize((256, 256), Image.ANTIALIAS)

# Save or display the resized image
resized_image.save('resized_image.png')  # Save the resized image as a PNG file
resized_image.show()  # Display the resized image using the default viewer
```

## Generate constellation

```python
import numpy as np
import matplotlib.pyplot as plt
# Generate binary data (0s and 1s)
binary_data = np.array([0,1])

# BPSK modulation: 0 -> -1, 1 -> 1
bpsk_symbols = 2 * binary_data - 1

# Plot BPSK constellation
plt.figure(figsize=(8, 8))
plt.scatter(bpsk_symbols.real, bpsk_symbols.imag, color='b', marker='o', label='BPSK Symbols')
plt.title('BPSK Constellation')
plt.xlabel('Real')
plt.ylabel('Imag')

plt.vlines(0, -1.5, 1.5, colors='k')
plt.hlines(0, -1.5, 1.5, colors='k')

# Set axis limits to center at (0, 0)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

# Add legend
plt.legend()

# Annotate point (-1, 0) as 'bit -1'
plt.annotate('bit 0', xy=(-1, 0), xytext=(-1., 0.1))
plt.annotate('bit 1', xy=(-1, 0), xytext=(1., 0.1))

plt.grid(True)
plt.show()
plt.savefig('./Images/bpsk_constellation.png', bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# Generate random binary data for QPSK
GrayMap = 1/np.sqrt(2) * np.array([(-1-1j), (-1+1j), (1-1j), (1+1j)])

# Plot QPSK constellation
fig, ax = plt.subplots(figsize=(8, 8))

ax.scatter(GrayMap.real, GrayMap.imag, color='b', marker='o', label='QPSK Symbols')
ax.set_title('Normalized QPSK Constellation')
ax.set_xlabel('Real')
ax.set_ylabel('Imag')

# Draw axes
ax.axhline(0, color='k')
ax.axvline(0, color='k')

# Set axis limits to center at (0, 0)
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

# Add legend
ax.legend()

# Add circle
circle = Circle((0, 0), 1, edgecolor='k', facecolor='none')
ax.add_patch(circle)

# Add annotations
annotations = {'00': (-1, -1), '01': (-1, 1), '11': (1, 1), '10': (1, -1)}
for text, point in annotations.items():
    ax.annotate(text, xy=point, xytext=(point[0] + 0.1, point[1] + 0.1), size=12)

# Add radius annotation
ax.annotate(r'$r = \sqrt{2}$', xy=(1, -1), xytext=(0.2, 0.5), size=12)

# Add arrow to show radius value
ax.arrow(0, 0, 1/np.sqrt(2), 1/np.sqrt(2), color='red', linestyle='dashed')

plt.grid(True)
plt.show()
```
